```python
import pandas as pd
import json
```

```python
# -----------------------------
# Load JSON File
# -----------------------------
with open("MultipledataSources2/Customers.json", "r") as f:
    customers_data = json.load(f)

customers_df = pd.DataFrame(customers_data)
```

```python
# -----------------------------
# Load CSV File
# -----------------------------
orders_df = pd.read_csv("MultipledataSources2/Orders.csv")
```

```python
# -----------------------------
# Load Excel File
# -----------------------------
payments_df = pd.read_excel("MultipledataSources2/Payments.xlsx")
```

```python
# -----------------------------
# Data Cleaning
# -----------------------------

# Clean column names
customers_df.columns = customers_df.columns.str.strip().str.lower()
orders_df.columns = orders_df.columns.str.strip().str.lower()
payments_df.columns = payments_df.columns.str.strip().str.lower()

# Trim whitespace from string columns
customers_df['name'] = customers_df['name'].str.strip()
customers_df['email'] = customers_df['email'].str.strip().str.lower()

# Ensure customer_id is integer
customers_df['customer_id'] = customers_df['customer_id'].astype(int)
orders_df['customer_id'] = orders_df['customer_id'].astype(int)
payments_df['customer_id'] = payments_df['customer_id'].astype(int)

# Remove duplicates
customers_df = customers_df.drop_duplicates()
orders_df = orders_df.drop_duplicates()
payments_df = payments_df.drop_duplicates()

# -----------------------------
# Merge Data
# -----------------------------

# Merge customers with orders
merged_df = pd.merge(customers_df, orders_df, 
                     on="customer_id", 
                     how="inner")

# Merge result with payments
final_df = pd.merge(merged_df, payments_df, 
                    on="customer_id", 
                    how="left")

# -----------------------------
# Output Result
# -----------------------------
print(final_df)

# Save final output
final_df.to_csv("final_merged_data.csv", index=False)
```
